# Loading Data

we will load all the csv files states[0-9] and save them into single dataframe  and will name it "us_census"

```python
us_census = pd.concat([pd.read_csv(f) for f in glob.glob('states[0-9].csv')])
```

```python
us_census.head(10)
```

```python
us_census.reset_index(drop=True, inplace=True)
```

```python
us_census.head(10)
```

Dropping the Unnamed column because its of no use

```python
us_census.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
us_census.columns
```

```python
us_census.dtypes
```

we will have to convert data types into numeric values for graph plotting.

```python
#removing $ sign from Income column

us_census['Income'].replace('\$', '', regex=True, inplace=True)
```

```python
us_census.tail()
```

We will add two new columns "Men" and "Women" by splitting 'GenderProp' column

```python
us_census['Men'], us_census['Women'] = us_census['GenderPop'].str.split('_', 1).str
```

```python
us_census.head()
```

Removing 'M' and 'F' from the newely created columns

```python
us_census['Men']= us_census['Men'].str.replace('M', '')
us_census['Women']= us_census['Women'].str.replace('F', '')
```

```python
us_census.head()
```

Converting new columns type to numeric

```python
us_census[["Income", "Women"]] = us_census[["Income", "Women"]]. apply(pd.to_numeric)
```

```python
us_census.dtypes
```

plotting scatterd graph

```python
x = us_census['Women']
y = us_census['Income']


plt.scatter(x, y)
plt.show()
```

Checking for empty values

```python
us_census.isnull().sum()
```

```python
us_census["Men"] = us_census["Men"]. apply(pd.to_numeric)
```

Filling empty values in "Women" column

```python
us_census['Women'].fillna(us_census['TotalPop'] - us_census['Men'], inplace=True)
```

```python
us_census['Women']
```

Checking for duplicates

```python
us_census.duplicated()
```

dropping dupltcated rows

```python
us_census.drop_duplicates(inplace=True)
```

```python
us_census.duplicated().sum()
```

Converting non-numeric data types to numeric type

```python
us_census = us_census.astype({"Hispanic": float, "White": float, "Black": float, "Native": float, "Asian": float, "Pacific": float})
```

```python
us_census.dtypes
```

Filling missing values in pacific Columns with mean of that colmun

# Population Distribution of first six states

```python
us_census.head(6)
```

```python
pop_dist = us_census[:6][['State','Hispanic','White','Black','Native','Asian','Pacific']]
```

```python
pop_dist
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, (idx, row) in enumerate(pop_dist.set_index('State').iterrows()):
    ax = axes[i // 3, i % 3]
    row = row[row.gt(row.sum() * .01)]
    ax.pie(row, labels=row.index, startangle=30)
    ax.set_title(idx)

fig.subplots_adjust(wspace=.2)
```

```python
tp_dist = us_census[:6][['State','TotalPop']]
```

```python
x = tp_dist['State']
y = tp_dist['TotalPop']


plt.bar(x, y, width = 0.8, color = ['red', 'green'])
  
plt.xlabel('State')
plt.ylabel('Total Population')
plt.title('Population Chart')
  
plt.show()
```
